```python
import cv2
from matplotlib import pyplot as plt
import os

# Replace 'sample_image.jpg' with the name of your uploaded image file
image_name = 'smileyFace.jpg'

if os.path.exists(image_name):
    img = cv2.imread(image_name)
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    plt.imshow(img_rgb)
    plt.title("Displayed Image using OpenCV and Matplotlib")
    plt.axis("off")
    plt.show()
else:
    print(f"The file '{image_name}' was not found. Please upload the image file.")
```
